4. Counting vowels in a given word

```python
vowels = ['a', 'e', 'i', 'o', 'u']
word = "Hello, World!"
vowels_in_word = [char for char in word if char.lower() in vowels]
count = 0
for char in word:
    if char.lower() in vowels:
        count += 1
print("Number of vowels in the string:", count)
print("Vowels in the string:", vowels_in_word)
```

5. Counting consonants in a given word

```python
vowels = ['a', 'e', 'i', 'o', 'u']
word = "Hello, World!"
consonants_in_word = [char for char in word if char.lower() not in vowels and char.isalpha()]
count = 0
for char in word:
    if char.lower() not in vowels and char.isalpha():
        count += 1
print("Number of consonants in the string:", count)
print("Consonants in the string:", consonants_in_word)
```

6. Counting the number of occurances of a character in a String

```python
word = "Hello, World!"
character = 'l'
count = 0
for char in word:
    if char == character:
        count += 1
print(f"The character '{character}' appears {count} times in the string.")
```

7. Fibonacci Series

```python
fib = [0,1]

n = 5
for i in range(2,n):
    next_fib = fib[i-1] + fib[i-2]
    fib.append(next_fib)
print(f"The first {n} Fibonacci numbers are: {fib}")
```

8. Finding the Maximum Number in a list 

```python
number_list = [15, 92, 23, 34, 65,34]
max_number = max(number_list)
min_number = min(number_list)
print(f"The maximum number in the list is: {max_number}")
print(f"The minimum number in the list is: {min_number}")
```

```python
number_list = [15, 92, 23, 34, 65,34]
max_number = number_list[0]
for number in number_list:
    if number > max_number:
        max_number = number
min_number = number_list[0]
for number in number_list:
    if number < min_number:
        min_number = number
print(f"The maximum number in the list is: {max_number}")
print(f"The minimum number in the list is: {min_number}")
```

12. Adding Two List Elements together

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

result_list = [a + b for a, b in zip(list1, list2)]
print(result_list)  # Output: [5, 7, 9]

res_list = []
for i in range(len(list1)):
    res_list.append(list1[i] + list2[i])
print(res_list)  # Output: [5, 7, 9]
```

13. Compare two Strings for anagrams

```python
str1 = "Listen"
str2 = "Silent"

if sorted(str1.lower()) == sorted(str2.lower()):
    print(f"{str1} and {str2} are anagrams.")
```

14. Checking for palindrome using extended slicing technique

```python
str1 = "Racecar"

if str1.lower() == str1.lower()[::-1]:
    print(f"{str1} is a palindrome.")
else:
    print(f"{str1} is not a palindrome.")
```

15. Counting Digits, Letters and Spaces in String

```python
import re

name = "John Doe 12"

digitCount = len(re.findall(r'\d', name))
letterCount = len(re.findall(r'[A-Za-z]', name))
spaceCount = len(re.findall(r'\s', name))

print(f"Number of digits in the name: {digitCount}")
print(f"Number of letters in the name: {letterCount}")
print(f"Number of spaces in the name: {spaceCount}")
```

16. Counting special characters in a string

```python
def count_special_characters(name):
    sp_characters = re.findall(r'[^A-Za-z0-9\s]', name)
    return len(sp_characters)

test_name = "John Doe!@#"
special_char_count = count_special_characters(test_name)
print(f"Number of special characters in the name: {special_char_count}")
```

18. Removing all whilespace in a string

```python
import re

string = "Hello, World! 123"
spaces = re.compile(r'\s+')
result = re.sub(spaces,"", string)

print(result)  # Output: Hello,World!123
```

```python
string = "Hello, World! 123 456" 
string2 = "".join(string.split())
print(string2)  # Output: Hello,World!123456
```

```python
string = "Hello, World! 123 456" 
string2 = string.replace(" ", "")
print(string2)  # Output: Hello,World!123456
```
