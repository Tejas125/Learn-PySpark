1. Introduction to Pandas

Pandas is a Python library for tabular data (rows + columns), similar to Excel or SQL tables but much more powerful.[youtube +1]

```python
import pandas as pd
```

Main objects:

•	 Series : 1D labeled array (like a single column).

•	 DataFrame : 2D labeled table (like an entire sheet).

2. Creating Pandas Series

```python
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
s
```

•	Values:  [10, 20, 30] 
•	Index:  ['a', 'b', 'c'] 

You can also create from a dict:

```python
s2 = pd.Series({'x': 1, 'y': 2, 'z': 3})
s2
```

Series of temperatures for 7 days with day names as index

```python
temps = pd.Series(
    [32, 34, 39, 41, 33, 40, 38],
    index=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
)
temps
```

3. Creating DataFrames From dictionaries

```python
data = {
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 28],
    'city': ['Ahilyanagar', 'Pune', 'Mumbai']
}
df = pd.DataFrame(data)
df
```

From list of lists / list of dicts

```python
# List of lists
rows = [
    ['Alice', 25, 'Ahilyanagar'],
    ['Bob', 30, 'Pune'],
    ['Charlie', 28, 'Mumbai']
]
df2 = pd.DataFrame(rows, columns=['name', 'age', 'city'])

# List of dicts
rows2 = [
    {'name': 'Alice', 'age': 25, 'city': 'Ahilyanagar'},
    {'name': 'Bob', 'age': 30, 'city': 'Pune'},
]
df3 = pd.DataFrame(rows2)
df3
```

From Series

```python
s_name = pd.Series(['Alice', 'Bob'], name='name')
s_age = pd.Series([25, 30], name='age')
df4 = pd.DataFrame({'name': s_name, 'age': s_age})
df4
```

DataFrame of students with columns  name ,  math_score ,  science_score  using a dict.

```python
students = {
    'name': ['Dave', 'Eve', 'Frank'],
    'math_score': [88, 92, 75],
    'science_score': [90, 85, 80]
}
df_students = pd.DataFrame(students)
print(df_students)
```
