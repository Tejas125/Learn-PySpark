3. Creating DataFrames From dictionaries

```python
data = {
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 28],
    'city': ['Ahilyanagar', 'Pune', 'Mumbai']
}
df = pd.DataFrame(data)
df
```

From list of lists / list of dicts

```python
# List of lists
rows = [
    ['Alice', 25, 'Ahilyanagar'],
    ['Bob', 30, 'Pune'],
    ['Charlie', 28, 'Mumbai']
]
df2 = pd.DataFrame(rows, columns=['name', 'age', 'city'])

# List of dicts
rows2 = [
    {'name': 'Alice', 'age': 25, 'city': 'Ahilyanagar'},
    {'name': 'Bob', 'age': 30, 'city': 'Pune'},
]
df3 = pd.DataFrame(rows2)
df3
```

5. Extracting Rows Using  loc  and  iloc 
Key difference:

•	 loc : label-based (use index/column names).

•	 iloc : position-based (use 0-based integer indices).

```python
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 28],
}, index=['r1', 'r2', 'r3'])
df
```

Using  loc  (labels)

```python
# Single row by label
df.loc['r2']

# Multiple rows
df.loc[['r1', 'r3']]

# Row and column by label
df.loc['r2', 'age']

# Slice (end inclusive)
df.loc['r1':'r3', 'name':'age']  # includes 'r3' and 'age'
```

Using  iloc  (positions)

```python
# Single row by position
df.iloc[1]  # second row

# Multiple rows
df.iloc[[0, 2]]

# Row and column by position
df.iloc[1, 1]  # second row, second column

# Slice (end exclusive)
df.iloc[0:2, 0:2]  # rows 0,1; cols 0,1 (end not included)
```

6. Pandas Dealing with Rows and Columns

Selecting columns

```python
df['name']          # Series
df[['name', 'age']] # DataFrame
```

Adding / modifying columns

```python
df['grade'] = ['A', 'B', 'C']
df['age_plus_1'] = df['age'] + 1
df
```

Deleting columns / rows

```python
df.drop('grade', axis=1)         # column
df.drop('r2', axis=0)            # row
df.drop(columns=['grade'])       # alternate
df.drop(index=['r2'])            # alternate
df
```

Renaming

```python
df.rename(columns={'name': 'student_name'},
          index={'r1': 'row1'})
df
```

9. Pandas Timedelta

Timedelta  represents durations (days, hours, etc.).

```python
td = pd.Timedelta(days=2, hours=3)
td
```

Use in DataFrames:

```python
df = pd.DataFrame({
    'start': pd.to_datetime(['2025-01-01', '2025-01-05']),
    'duration': pd.to_timedelta(['1 day', '2 days'])
})
df['end'] = df['start'] + df['duration']
```
