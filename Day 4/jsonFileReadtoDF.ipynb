```python
import json
import pandas as pd
with open('data.json') as f:
    data = json.load(f)

# Normalizes nested JSON into a flat table
df = pd.json_normalize(data)
```

```python
df
```

```python
df.columns = df.columns.str.lower()
```

Apply Data Cleaning Technique (Normalize and Merge)

```python
df_cleaned = df.groupby(df.columns, axis=1).first()

print("Original DataFrame:\n", df)
print("\nCleaned DataFrame:\n", df_cleaned)
```
